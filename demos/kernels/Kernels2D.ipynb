#  Demos for how to get TGD kenels for two-dimentional usage

```python
# get the TGD lib
! pip install -i https://test.pypi.org/simple/ TGD==0.9.1

from TGD import TGDkernel

import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np
```

## Two-dimensional first-order derivatives TGD by Orthogonal construction method

```python
radius = 7
semi_width = 7
theta = 0
gauss_param = {'delta': 2.4}

tgd = TGDkernel()
```

```python
xv,yv, kernel = tgd.getKernel2D(radius= radius, 
                                method = 'directional',
                                kernel_size=141, 
                                theta = theta, 
                                paramx = gauss_param, paramy = gauss_param, 
                                kernel_func = 'Gaussian')
print(xv.shape,yv.shape, kernel.shape)
print(np.min(kernel),np.max(kernel))
```

```python
fig = plt.figure(figsize=(4, 4), dpi=100)
ax = fig.add_axes(Axes3D(fig))
surf = ax.plot_surface(xv,yv, kernel,
                       rstride=1,  # rstriderow
                       cstride=1,  # cstride(column)=
                       cmap='rainbow',
                       )
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

fig.colorbar(surf, shrink=0.5, aspect=5)
plt.show()
```

## Two-dimensional second-order derivatives TGD by Orthogonal construction method

```python
radius = 7
kernel_size=141
gauss_param = {'delta': 2.4}
xv,yv, kernel = tgd.getKernel2D(radius= radius, 
                                method = 'directional',
                                kernel_size=kernel_size, 
                                theta = theta, 
                                derivation_order = 2,
                                paramx = gauss_param, paramy = gauss_param, 
                                kernel_func = 'Gaussian')
print(xv.shape,yv.shape, kernel.shape)
print(np.min(kernel),np.max(kernel))
print(kernel.sum())
```

## Two-dimensional first-order derivatives TGD by Rotational construction method

```python
radius = 7
kernel_size= 141
theta = 0
gauss_param = {'delta': 2.4}
exp_param = {'k': 1, 'delta': 0.56}
xv,yv, kernel = tgd.getKernel2D(radius= radius, 
                                method = 'rotational',
                                kernel_size=kernel_size, 
                                theta = theta, 
                                derivation_order = 1,
                                paramx = gauss_param, paramy = gauss_param, 
                                kernel_func = 'Gaussian'
                                # paramx = exp_param, paramy = exp_param, 
                                # kernel_func = 'Exp'
                                )
print(xv.shape,yv.shape, kernel.shape)
print(np.min(kernel),np.max(kernel))
```

```python
fig = plt.figure(figsize=(4, 4), dpi=100)
ax = fig.add_axes(Axes3D(fig)) 
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
surf = ax.plot_surface(xv,yv, kernel, rstride = 1, cstride = 1, cmap='rainbow')
fig.colorbar(surf, shrink=0.5, aspect=10)
plt.show()
```
