```python
from TGD import TGDkernel
import matplotlib.pyplot as plt
```

```python
# get the TGD kernel
tgd = TGDkernel()
```

```python
# prepare parametere
radius = 20
gauss_param = {'delta': 6.6}
exp_param = {'k': 1, 'delta': 0.17}
linear_param = {'k': 2, 'b': radius * 2 + 0.4}
landau_param = {'c':7/16, 'n':3, 'r':radius+1}
weibull_param = {'k':1, 'lambda':6}
```

# One-dimentional first-order and second-order derivative TGD kernels  

##  Gaussian TGD kernel

```python
x1,kernel1 = tgd.getKernel1D(derivation_order=1, radius = 20, kernel_size= 401, param=gauss_param, kernel_func='Gaussian')
x2,kernel2 = tgd.getKernel1D(derivation_order=2, radius = 20, kernel_size= 401, param=gauss_param, kernel_func='Gaussian')

plt.figure(figsize=(20, 10), dpi=100)

plt.subplot(1,2,1)
ax=plt.gca() 
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.spines['bottom'].set_position(('data',0)) 
ax.spines['left'].set_position(('axes',0.5))
plt.plot(x1,kernel1)
plt.title('First-order 1D Gaussian TGD kernel')
plt.subplot(1,2,2)
ax=plt.gca() 
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.spines['bottom'].set_position(('data',0)) 
ax.spines['left'].set_position(('axes',0.5))
plt.title('Second-order 1D Gaussian TGD kernel')
plt.plot(x2,kernel2)
        
plt.subplots_adjust(left=None, bottom=None, right=None, top=None,
                wspace=0, hspace=0)
```
